```python
import pandas as pd

# Load the datasets
customers = pd.read_csv('C:/Users/shara/OneDrive/Desktop/Zeotap/Customers.csv')
products = pd.read_csv('C:/Users/shara/OneDrive/Desktop/Zeotap/Products.csv')
transactions = pd.read_csv('C:/Users/shara/OneDrive/Desktop/Zeotap/Transactions.csv')

# Display the first few rows of each dataset
print(customers.head())
print(products.head())
print(transactions.head())
```

```python
# Check for missing values
print(customers.isnull().sum())
print(products.isnull().sum())
print(transactions.isnull().sum())

# Check for duplicates
print(customers.duplicated().sum())
print(products.duplicated().sum())
print(transactions.duplicated().sum())
```

```python
import matplotlib.pyplot as plt

# Convert SignupDate to datetime
customers['SignupDate'] = pd.to_datetime(customers['SignupDate'])

# Plot customer signups over time
customers['SignupDate'].value_counts().sort_index().plot(kind='line')
plt.title('Customer Signups Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Signups')
plt.show()
```

```python
# Plot the distribution of product categories
products['Category'].value_counts().plot(kind='bar')
plt.title('Product Category Distribution')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the distribution of transaction values
transactions['TotalValue'].plot(kind='hist', bins=50)
plt.title('Distribution of Transaction Values')
plt.xlabel('Total Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Descriptive statistics for customers
print(customers.describe())

# Descriptive statistics for products
print(products.describe())

# Descriptive statistics for transactions
print(transactions.describe())
```
